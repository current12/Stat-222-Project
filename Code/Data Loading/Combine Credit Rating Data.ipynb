# Combine Credit Rating Data

Original data source: https://www.kaggle.com/datasets/agewerc/corporate-credit-rating 

Supplementary data source: https://www.kaggle.com/datasets/kirtandelwadia/corporate-credit-rating-with-financial-ratios

```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load original data into pandas
# ~\..\Box\STAT 222 Capstone\Raw Data\Credit Rating Data From Kaggle\corporate_rating.csv
corporate_rating = pd.read_csv('~\Box\STAT 222 Capstone\Raw Data\Credit Rating Data From Kaggle\corporate_rating.csv')
corporate_rating
```

```python
# Load supplementary data
# C:\Users\ijyli\Box\STAT 222 Capstone\Raw Data\Supplementary Credit Rating Data From Kaggle\corporateCreditRatingWithFinancialRatios.csv
supplementary_data = pd.read_csv('~\Box\STAT 222 Capstone\Raw Data\Supplementary Credit Rating Data From Kaggle\corporateCreditRatingWithFinancialRatios.csv')
supplementary_data
```

```python
# All columns
print(corporate_rating.columns)
print(supplementary_data.columns)
```

```python
# In original data, limit to Rating, Name, Symbol, Rating Agency Name, Date, Sector
corporate_rating = corporate_rating[['Rating', 'Name', 'Symbol', 'Rating Agency Name', 'Date', 'Sector']]
# Create Source column as 'Original'
corporate_rating['Source'] = 'Original'
# Convert Date to date format
corporate_rating['Date'] = pd.to_datetime(corporate_rating['Date'])
# In supplementary data, limit to Rating Agency, Corporation, Rating, Rating Date, Sector, Ticker
supplementary_data = supplementary_data[['Rating Agency', 'Corporation', 'Rating', 'Rating Date', 'Sector', 'Ticker']]
# Convert Rating Date to date format
supplementary_data['Rating Date'] = pd.to_datetime(supplementary_data['Rating Date'])
# Create Source column as 'Supplementary'
supplementary_data['Source'] = 'Supplementary'
# Rename columns in supplementary data to match original data
supplementary_data = supplementary_data.rename(columns = {'Rating Agency':'Rating Agency Name', 'Corporation':'Name', 'Rating Date':'Date', 'Ticker':'Symbol'})
# Combined data
combined_data = pd.concat([corporate_rating, supplementary_data], ignore_index = True)
combined_data
```

```python
# Missing values?
combined_data.isnull().sum()
```

```python
# Name column seems tricky to harmonize, so drop for now
combined_data = combined_data.drop(columns = ['Name'])
```

```python
# Count values of rating
print(combined_data['Rating'].value_counts())
# Strip + or - from Rating
combined_data['Rating'] = combined_data['Rating'].str.strip('+').str.strip('-')
# Count again
print(combined_data['Rating'].value_counts())
```

```python
# Count values of Rating Agency Name
print(combined_data['Rating Agency Name'].value_counts())
```

```python
# Count values of Sector
print(combined_data['Sector'].value_counts())
# Seems tricky to harmonize, so drop for now
combined_data = combined_data.drop(columns = ['Sector'])
```

```python
# Remaining columns
print(combined_data.columns)
```

```python
# Drop duplicates
combined_data = combined_data.drop_duplicates()
```

```python
# Number of rows
print(combined_data.shape[0])
# String collapse on 'Rating', 'Symbol', 'Rating Agency Name', 'Date'
# Combine values of Source column
combined_data = combined_data.groupby(['Rating', 'Symbol', 'Rating Agency Name', 'Date']).agg(', '.join).reset_index()
# Number of rows
print(combined_data.shape[0])
```

```python
# Values of Source
print(combined_data['Source'].value_counts())
# Recode 'Original, Supplementary' as 'Both'
combined_data['Source'] = combined_data['Source'].str.replace('Original, Supplementary', 'Both')
# Values of Source
print(combined_data['Source'].value_counts())
```

```python
# Summarize variables
combined_data.describe()
```

```python
# Print out value counts for each column
for col in combined_data.columns:
    print(col, combined_data[col].value_counts())
```

```python
# Encode Ratings as ordinal
rating_order = ['AAA', 'AA', 'A', 'BBB', 'BB', 'B', 'CCC', 'CC', 'C', 'D']
# Reverse
rating_order = rating_order[::-1]
# Create dictionary mapping to number values
rating_dict = {rating: i for i, rating in enumerate(rating_order, 1)}
# Create new column "Rating Rank AAA is 10"
combined_data['Rating Rank AAA is 10'] = combined_data['Rating'].map(rating_dict)

# Cross tab Rating and 'Rating Rank AAA is 10'
pd.crosstab(combined_data['Rating'], combined_data['Rating Rank AAA is 10'])
```

```python
# Variable for next rating, previous rating, and associated dates

# Sort appropriately
combined_data = combined_data.sort_values(by=['Rating Agency Name', 'Symbol', 'Date'])

# Next rating is next item if same Symbol and Rating Agency Name
combined_data['Next Rating'] = combined_data['Rating'].shift(-1)
# Set to NaN if next Symbol or Rating Agency Name are different
combined_data.loc[combined_data['Symbol'].shift(-1) != combined_data['Symbol'], 'Next Rating'] = None
combined_data.loc[combined_data['Rating Agency Name'].shift(-1) != combined_data['Rating Agency Name'], 'Next Rating'] = None

# Next rating date is next item if same Symbol and Rating Agency Name
combined_data['Next Rating Date'] = combined_data['Date'].shift(-1)
# Set to NaN if next Symbol or Rating Agency Name are different
combined_data.loc[combined_data['Symbol'].shift(-1) != combined_data['Symbol'], 'Next Rating Date'] = None
combined_data.loc[combined_data['Rating Agency Name'].shift(-1) != combined_data['Rating Agency Name'], 'Next Rating Date'] = None

# Previous rating is previous item if same Symbol and Rating Agency Name
combined_data['Previous Rating'] = combined_data['Rating'].shift(1)
# Set to NaN if previous Symbol or Rating Agency Name are different
combined_data.loc[combined_data['Symbol'].shift(1) != combined_data['Symbol'], 'Previous Rating'] = None
combined_data.loc[combined_data['Rating Agency Name'].shift(1) != combined_data['Rating Agency Name'], 'Previous Rating'] = None

# Previous rating date is previous item if same Symbol and Rating Agency Name
combined_data['Previous Rating Date'] = combined_data['Date'].shift(1)
# Set to NaN if previous Symbol or Rating Agency Name are different
combined_data.loc[combined_data['Symbol'].shift(1) != combined_data['Symbol'], 'Previous Rating Date'] = None
combined_data.loc[combined_data['Rating Agency Name'].shift(1) != combined_data['Rating Agency Name'], 'Previous Rating Date'] = None
```

```python
# Variable that is Next Rating Date or End of Data - 12/31/2016 if missing
combined_data['Next Rating Date or End of Data'] = combined_data['Next Rating Date']
# Set to 12/31/2016 if missing
combined_data.loc[combined_data['Next Rating Date'].isnull(), 'Next Rating Date or End of Data'] = pd.to_datetime('12/31/2016')
```

```python
# Mark changes in ratings
# Order by Rating Agency Name, Symbol, and Date
# If Rating is not equal to the lag but Rating Agency Name and Symbol are, flag the observation as a change in rating
combined_data = combined_data.sort_values(by=['Rating Agency Name', 'Symbol', 'Date'])
combined_data['Rating Change'] = (combined_data['Rating'] != combined_data['Rating'].shift(1)) & (combined_data['Rating Agency Name'] == combined_data['Rating Agency Name'].shift(1)) & (combined_data['Symbol'] == combined_data['Symbol'].shift(1))
# Mark item as ambiguous if rating agency or symbol changes
combined_data['Ambiguous'] = (combined_data['Rating Agency Name'] != combined_data['Rating Agency Name'].shift(1)) | (combined_data['Symbol'] != combined_data['Symbol'].shift(1))

# Create Type variable
# Ambiguous if Ambiguous is True
# Change if Rating Change is True
# Affirmation if both are False
combined_data['Type'] = 'Affirmation'
combined_data.loc[combined_data['Rating Change'], 'Type'] = 'Change'
combined_data.loc[combined_data['Ambiguous'], 'Type'] = 'Ambiguous'

# If Type is 'Change' and 'Rating Rank AAA is 10' is higher than the prior observation, flag as an upgrade
combined_data['Upgrade'] = (combined_data['Type'] == 'Change') & (combined_data['Rating Rank AAA is 10'] > combined_data['Rating Rank AAA is 10'].shift(1))
# If Type is 'Change' and 'Rating Rank AAA is 10' is lower than the prior observation, flag as a downgrade
combined_data['Downgrade'] = (combined_data['Type'] == 'Change') & (combined_data['Rating Rank AAA is 10'] < combined_data['Rating Rank AAA is 10'].shift(1))
# For Upgrades, set Type to 'Upgrade'
combined_data.loc[combined_data['Upgrade'], 'Type'] = 'Upgrade'
# For Downgrades, set Type to 'Downgrade'
combined_data.loc[combined_data['Downgrade'], 'Type'] = 'Downgrade'
# Drop Upgrade and Downgrade columns
combined_data = combined_data.drop(columns=['Upgrade', 'Downgrade'])

# Cross tabulate Type and Rating Change and Ambiguous
print(pd.crosstab(combined_data['Type'], combined_data['Rating Change']))
print(pd.crosstab(combined_data['Type'], combined_data['Ambiguous']))

# Drop Rating Change and Ambiguous columns
combined_data = combined_data.drop(columns=['Rating Change', 'Ambiguous'])

# Print 100 rows to check
print(combined_data.head(100))
```

```python
# Create change in rating variable
# Set to NaN by default
combined_data['Change in Rating'] = None
# If Type is Affirmation, set to 0
combined_data.loc[combined_data['Type'] == 'Affirmation', 'Change in Rating'] = 0
# If Type if Upgrade or Downgrade, set to 'Rating Rank AAA is 10' - 'Rating Rank AAA is 10' of the prior observation
combined_data.loc[combined_data['Type'] == 'Upgrade', 'Change in Rating'] = combined_data['Rating Rank AAA is 10'] - combined_data['Rating Rank AAA is 10'].shift(1)
combined_data.loc[combined_data['Type'] == 'Downgrade', 'Change in Rating'] = combined_data['Rating Rank AAA is 10'] - combined_data['Rating Rank AAA is 10'].shift(1)

# Print 100 rows to check
print(combined_data[['Rating Agency Name', 'Symbol', 'Rating Rank AAA is 10', 'Rating', 'Change in Rating']].head(100))
```

```python
# Sort by Rating Agency Name, Symbol, then Date
combined_data = combined_data.sort_values(by = ['Rating Agency Name', 'Symbol', 'Date'])
```
