```python
import pandas as pd
import numpy as np
import os
```

```python
nlp_new = pd.read_parquet("nlp_features_new.parquet")
nlp_new
```

```python
word_tone_new = pd.read_parquet("word_tone_new.parquet")
word_tone_new
```

```python
nlp_new = pd.concat([nlp_new, word_tone_new], axis=1)
```

```python
# Get new finbert data
path = '../../../Data/Finbert_Pos_Score'

# Get Finbert positivity score
# Combine finbert data
finbert_list = ['pos_result_new_0_400.txt',
                'pos_result_new_400_800.txt',
                'pos_result_new_800_1500.txt',]

finbert_list = [f"{path}/{i}" for i in finbert_list]

pos_score_fin = np.array([])

for filename in finbert_list:
    temp = np.loadtxt(filename)
    pos_score_fin = np.concatenate((pos_score_fin, temp))
    
print("Bad call index: ", np.where(pos_score_fin == -1))

nlp_new["pos_score_finbert"] = pos_score_fin
```

```python
# Combine with old data
file_list = [f for f in os.listdir(r'../../../Data/All_Data/All_Data_with_NLP_Features') if f.endswith('.parquet')]
# read in all nlp files
df_nlp = pd.concat([pd.read_parquet(r'../../../Data/All_Data/All_Data_with_NLP_Features/' + f) for f in file_list])
df_nlp
```

```python
df_nlp_new = pd.concat([df_added, nlp_new.loc[:, "num_transparency":]], axis=1)
```

```python
df_nlp_new = df_nlp_new.rename(columns={"readability": "gf_score"})
df_nlp_new = df_nlp_new.drop("index", axis=1)
```

```python
df_full = pd.concat([df_nlp.loc[:,df_nlp_new.columns], df_nlp_new], axis=0)
```

```python
df_full = df_full[df_full["pos_score_finbert"] != -1]    # Drop bad finbert

# Tone outlier
df_full = df_full.rename(columns={"TONE1": "tone"})

df_full = df_full.reset_index()
# Check tone outlier
idx = df_full['tone'].idxmax()

# Retrieve the row
max_row = df_full.loc[idx]
#print(df.loc[idx]['transcript'])

# Remove tone outlier
df_full = df_full.drop(labels=idx)
```

```python
# Get num_question / call length
df_full['num_q_by_len'] = df_full['num_questions']/df_full['word_count']
df_full['len_by_q_count'] = df_full['word_count']/df_full['num_questions']

df_full.drop(["level_0", "index"], axis=1).to_parquet("../../../Data/All_Data/NLP_Finbert_only.parquet")
# inf values in len_by_q_count as q_count=0 for calls w/out Q&A sess
#df.describe()
```
