```python
import pandas as pd
```

```python
df = pd.read_parquet('../../../../../Downloads/NLP_Finbert_only.parquet')
df
```

```python
# merge on fixed_quarter_date

# list of files in '../../../Data/All_Data/All_Data_Fixed_Quarter_Dates'
import os
file_list = [f for f in os.listdir(r'../../../Data/All_Data/All_Data_Fixed_Quarter_Dates') if f.endswith('.parquet')]
# read in all parquet files
all_data_fqd_df = pd.concat([pd.read_parquet(r'../../../Data/All_Data/All_Data_Fixed_Quarter_Dates/' + f, columns = ['ticker', 'fixed_quarter_date', 'transcript']) for f in file_list])

df = df.merge(all_data_fqd_df, on = ['ticker', 'transcript'], how = 'left')
df
```

```python
# check columns
print(df.columns)
```

```python
df.drop(columns = ['transcript'], inplace=True)
```

```python
df.to_parquet('../../../Data/NLP Only/NLP_Finbert_only.parquet')
```
