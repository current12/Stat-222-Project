# Combine Original and Filled Sector Data and Screener Data

Combine data sources

```python
# Packages
import pandas as pd
```

## Original Data

```python
# US Stocks data
# '~\Box\STAT 222 Capstone\Raw Data\US Public Companies Classification\US_stocks.csv'
us_stocks = pd.read_csv('~/Box/STAT 222 Capstone/Raw Data/US Public Companies Classification/US_stocks.csv')
us_stocks
```

## Filled Data

```python
# Filled in values
# '~\Box\STAT 222 Capstone\Intermediate Data\Sectors\Missing_in_sector_data - Filled.csv'
filled = pd.read_csv('~/Box/STAT 222 Capstone/Intermediate Data/Sectors/Missing_in_sector_data - Filled.csv')
filled
```

```python
# Rename Symbol to Ticker
filled = filled.rename(columns={'Symbol': 'Ticker'})
```

## Combine Original and Filled

```python
# Combine the two dataframes by stacking them
combined = pd.concat([us_stocks, filled], axis=0)
combined
```

## Screener Data

```python
# Load '~\Box\STAT 222 Capstone\Raw Data\Screener\nasdaq_screener_1711519842163.csv'
nasdaq_screener = pd.read_csv('~/Box/STAT 222 Capstone/Raw Data/Screener/nasdaq_screener_1711519842163.csv')
nasdaq_screener
```

```python
# Keep Symbol, Name, Country, IPO Year, Sector, Industry
nasdaq_screener = nasdaq_screener[['Symbol', 'Name', 'Country', 'IPO Year', 'Sector', 'Industry']]
# Rename Sector to Screener_Sector, Industry to Screener_Industry, Symbol to Ticker
nasdaq_screener = nasdaq_screener.rename(columns={'Sector': 'Screener_Sector', 'Industry': 'Screener_Industry', 'Symbol': 'Ticker'})
# Merge combined and nasdaq_screener on Ticker
combined = pd.merge(combined, nasdaq_screener, on='Ticker', how='left')
combined
```

```python
# Count NaN values in Screener_Sector and Screener_Industry
print('obs missing Screener_Sector:', combined['Screener_Sector'].isna().sum())
print('obs missing Screener_Industry:', combined['Screener_Industry'].isna().sum())
# Count obs missing Sub-Industry
print('obs missing Sub-Industry:', combined['Sub-Industry'].isna().sum())
```

Screener is missing a ton of companies, so not very useful. The screener sector values also seem inaccurate.

## Limit variables to just Ticker and Sector

```python
combined = combined[['Ticker', 'Sector']]
# Check for missing values
print('obs missing Sector:', combined['Sector'].isna().sum())
combined
```

## Final Output

```python
# Output
# '~\Box\STAT 222 Capstone\Intermediate Data\Sectors\combined_sector_data.csv'
combined.to_csv('~/Box/STAT 222 Capstone/Intermediate Data/Sectors/combined_sector_data.csv', index=False)
```
