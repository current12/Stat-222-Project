# Combine Original and Filled Sector Data

Combine data sources

```python
# Packages
import pandas as pd
```

```python
# US Stocks data
# '~\Box\STAT 222 Capstone\Raw Data\US Public Companies Classification\US_stocks.csv'
us_stocks = pd.read_csv('~/Box/STAT 222 Capstone/Raw Data/US Public Companies Classification/US_stocks.csv')
us_stocks
```

```python
# Filled in values
# '~\Box\STAT 222 Capstone\Intermediate Data\Sectors\Missing_in_sector_data - Filled.csv'
filled = pd.read_csv('~/Box/STAT 222 Capstone/Intermediate Data/Sectors/Missing_in_sector_data - Filled.csv')
filled
```

```python
# Rename Symbol to Ticker
filled = filled.rename(columns={'Symbol': 'Ticker'})
```

```python
# Combine the two dataframes by stacking them
combined = pd.concat([us_stocks, filled], axis=0)
combined
```

```python
# Output
# '~\Box\STAT 222 Capstone\Intermediate Data\Sectors\combined_sector_data.csv'
combined.to_csv('~/Box/STAT 222 Capstone/Intermediate Data/Sectors/combined_sector_data.csv', index=False)
```
