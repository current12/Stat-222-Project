# Missing Sector Data

Find companies with credit rating data missing GICS sector data

```python
# Packages
import pandas as pd
```

```python
# Load combined credit rating data and keep list of unique companies
combined_credit_rating_data = pd.read_csv('~\Box\STAT 222 Capstone\Intermediate Data\Credit_Rating\combined_credit_rating_data.csv')
combined_credit_rating_data
```

```python
# Unique values of Symbol
unique_symbols = combined_credit_rating_data[['Symbol']].drop_duplicates()
unique_symbols
```

```python
# Load CSV 'C:\Users\ijyli\Box\STAT 222 Capstone\Raw Data\US Public Companies Classification\US_stocks.csv'
us_stocks = pd.read_csv(r'~\Box\STAT 222 Capstone\Raw Data\US Public Companies Classification\US_stocks.csv')
print(us_stocks)
```

```python
# Left join unique_symbols with us_stocks on 'Symbol' = 'Ticker'
companies_and_sectors = pd.merge(unique_symbols, us_stocks, left_on='Symbol', right_on='Ticker', how='left')
companies_and_sectors
```

```python
# Print items missing in sector data 
missing_in_sector_data = companies_and_sectors[companies_and_sectors['Ticker'].isnull()]
print(missing_in_sector_data)
```

```python
# Output value_counts of Sector to see items to choose from
print(us_stocks['Sector'].value_counts())
```

```python
# Output to CSV
# '~\Box\STAT 222 Capstone\Intermediate Data\Sectors\Missing_in_sector_data.csv'
missing_in_sector_data[['Symbol', 'Sector']].to_csv(r'~\Box\STAT 222 Capstone\Intermediate Data\Sectors\Missing_in_sector_data.csv', index=False)
```
