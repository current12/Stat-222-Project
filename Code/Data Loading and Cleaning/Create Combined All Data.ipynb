# Create Combined All Data

```python
# Whether this is a sample/debug run or not
DEBUG = False
```

```python
# Packages
import pandas as pd
import openpyxl
```

```python
# Load ~\Box\STAT 222 Capstone\Intermediate Data\credit_ratings_on_earnings_call_dates.csv
credit_ratings_on_earnings_call_dates = pd.read_csv(r'~\Box\STAT 222 Capstone\Intermediate Data\credit_ratings_on_earnings_call_dates.csv')
credit_ratings_on_earnings_call_dates
```

```python
# Load ~\Box\STAT 222 Capstone\Intermediate Data\calls.csv
if DEBUG:
    calls = pd.read_csv(r'~\Box\STAT 222 Capstone\Intermediate Data\calls.csv', nrows = 10)
else:
    calls = pd.read_csv(r'~\Box\STAT 222 Capstone\Intermediate Data\calls.csv')
# Delete the first column (index)
calls = calls.iloc[:,1:]
# Convert date to be just the date part
calls['date'] = calls['date'].str[:10]
calls
```

```python
# Inner join credit_ratings_on_earnings_call_dates.csv (on columns ticker, earnings_call_date) with calls.csv (on columns company, date)
merged = pd.merge(credit_ratings_on_earnings_call_dates, calls, left_on=['ticker', 'earnings_call_date'], right_on=['company', 'date'], how='inner')
merged

# Print range of dates
print('date range')
print(merged['date'].min())
print(merged['date'].max())

# Range of year, Year
for variable in ['year', 'Year']:
    print(variable)
    print(merged[variable].min())
    print(merged[variable].max())

# Print rows where year is greater than 2016
# This is due to inconsistencies in the earnings call data year variable
print('years greater than 2016')
print(merged[merged['year'] > 2016][['earnings_call_date', 'ticker', 'company', 'date', 'year', 'Year']])

# Free up memory from calls
del calls
```

```python
# Inner join this with the three tabuler_fin_data files (on columns company, year, quarter from earnings calls and columns symbol, calendarYear, period (with Q removed) from fin data)

files_to_load = [r'balance_sheet', r'cash_flow_statement', r'income_statement']

# Loop through and inner join each file onto merged
for file in files_to_load:
    print('loading ' + file)
    df = pd.read_csv(r'~\Box\STAT 222 Capstone\Intermediate Data\tabuler_fin_data(' + file + r').csv')
    # Delete the first column (index)
    df = df.iloc[:,1:]
    # Remove Q from period
    df['period'] = df['period'].str.replace('Q', '')
    # Convert to int
    df['period'] = df['period'].astype(int)
    # Print values of Period
    print(df['period'].unique())
    merged = pd.merge(merged, df, left_on=['company', 'year', 'quarter'], right_on=['symbol', 'calendarYear', 'period'], how='inner')
    # Remove suffix _x
    merged.columns = merged.columns.str.replace('_x', '')
    # Harmonize suffix y
    merged.columns = merged.columns.str.replace('_y', '_' + file)
    del df
    
merged
```

```python
# Rating change variables
# Get numeric values for Rating and rating_on_next_earnings_call_date
# Encode Ratings as ordinal
rating_order = ['AAA', 'AA', 'A', 'BBB', 'BB', 'B', 'CCC', 'CC', 'C', 'D']
# Reverse
rating_order = rating_order[::-1]
# Create dictionary mapping to number values
rating_dict = {rating: i for i, rating in enumerate(rating_order, 1)}
# Create new column "Rating Rank AAA is 10"
merged['Rating Rank AAA is 10'] = merged['Rating'].map(rating_dict)
# Create new column "Rating on Next Earnings Call Date Rank AAA is 10"
merged['Rating on Next Earnings Call Date Rank AAA is 10'] = merged['rating_on_next_earnings_call_date'].map(rating_dict)

# Cross tab Rating and 'Rating Rank AAA is 10'
print(pd.crosstab(merged['Rating'], merged['Rating Rank AAA is 10']))
# Cross tab rating_on_next_earnings_call_date and 'Rating on Next Earnings Call Date Rank AAA is 10'
print(pd.crosstab(merged['rating_on_next_earnings_call_date'], merged['Rating on Next Earnings Call Date Rank AAA is 10']))
```

```python
# Variable for change before next call
def change(row):
    if row['Rating Rank AAA is 10'] > row['Rating on Next Earnings Call Date Rank AAA is 10']:
        return 'Downgrade Before Next Call'
    elif row['Rating Rank AAA is 10'] < row['Rating on Next Earnings Call Date Rank AAA is 10']:
        return 'Upgrade Before Next Call'
    elif row['Rating Rank AAA is 10'] == row['Rating on Next Earnings Call Date Rank AAA is 10']:
        return 'Same At Next Call'
    else:
        return None

# Apply function to create new column
merged['Change Direction Before Next Call'] = merged.apply(change, axis = 1)

# Also numeric version 'Change Before Next Call'
merged['Change Before Next Call'] = merged['Rating on Next Earnings Call Date Rank AAA is 10'] - merged['Rating Rank AAA is 10']

# Cross tab 'Change Direction Before Next Call' and 'Change Before Next Call'
print(pd.crosstab(merged['Change Direction Before Next Call'], merged['Change Before Next Call']))
```

```python
# Print range of dates
print('date range')
print(merged['date'].min())
print(merged['date'].max())

# Range of year, Year, calendarYear
for variable in ['year', 'Year', 'calendarYear']:
    print(variable)
    print(merged[variable].min())
    print(merged[variable].max())
```
