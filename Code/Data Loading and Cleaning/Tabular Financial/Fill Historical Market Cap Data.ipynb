# Fill Historical Market Cap Data

Get market cap data to a daily level

```python
# Packages
import pandas as pd
```

```python
# Read in the data
# C:\Users\ijyli\Box\STAT 222 Capstone\Raw Data\Market Cap Data\market_cap_data.parquet
market_cap_data = pd.read_parquet('~/Box/STAT 222 Capstone/Raw Data/Market Cap Data/market_cap_data.parquet')
# Convert date to datetime
market_cap_data['date'] = pd.to_datetime(market_cap_data['date'])
market_cap_data
```

```python
# Create dataset of every symbol and every day from 2010-01-01 to 2016-12-31
# Create a list of dates from 2010-01-01 to 2016-12-31
dates = pd.date_range(start='2010-01-01', end='2016-12-31', freq='D')
symbols = market_cap_data['symbol'].unique()
# multi-index
index = pd.MultiIndex.from_product([symbols, dates], names=['symbol', 'date'])
# Create a dataframe with the multi-index
dates = pd.DataFrame(index=index).reset_index()
dates
```

```python
# Outer join dataframes on date
daily_market_cap = pd.merge(dates, market_cap_data, how='outer', on=['symbol', 'date'])
daily_market_cap
```

```python
# Forward fill marketCap by symbol
daily_market_cap['marketCap'] = daily_market_cap.groupby('symbol')['marketCap'].ffill()
daily_market_cap
```

```python
# Drop still-NaN items
daily_market_cap = daily_market_cap.dropna()
daily_market_cap
```

```python
# Output to parquet
daily_market_cap.to_parquet('~\Box\STAT 222 Capstone\Intermediate Data\Tabular_Fin/daily_market_cap.parquet')
```
