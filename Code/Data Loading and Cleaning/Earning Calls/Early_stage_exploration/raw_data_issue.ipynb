```python
# Load Retail/semiconductor/software
calls = {}

num_retail = 0
num_semiconductors = 0
num_software = 0
path = os.path.expanduser("~/Box/STAT 222 Capstone/Raw Data/Big Earnings Call Dataset/sectors/")
for sector in ["retail", "semiconductors", "software"]:
    print("Start loading:", sector)
    sector_path = os.path.join(path, sector)
    calls[sector] = {}
    for company in os.listdir(sector_path):
        company_path = os.path.join(sector_path, company)
        for call in os.listdir(company_path):
            # skip speaker name file
            if call == "speaker names.csv":
                continue
            call_path = os.path.join(company_path, call)

            # Note the full number of raw data
            if sector == "retail":
                num_retail += 1
            elif sector == "semiconductors":
                num_semiconductors += 1
            elif sector == "software":
                num_software += 1

            try:
                calls[sector][(company, call.strip(".csv")[-5:])] = pd.read_csv(call_path).T[[1,2]]
            except Exception as e:
                print(call_path)
                continue

# This Cell takes a hour to run
```

```python
print(f"Raw retail: {num_retail}")
print(f"Raw semiconductors: {num_semiconductors}")
print(f"Raw software: {num_software}")
print(f"Loadable retail: {len(calls["retail"])}")
print(f"Loadable semiconductors: {len(calls["semiconductors"])}")
print(f"Loadable software: {len(calls["software"])}")
```
