## Convert to Dataframe

```python
# Load data (need to get cat values)
df = load_data()

# Get the column names and mapping
numeric_feature_columns, cat_feature_columns, target_column, custom_mapping = get_column_names_and_mapping(model_name)

# Get feats
_, _, _, _, feats = prepare_matrices(df, numeric_feature_columns, cat_feature_columns, target_column, custom_mapping)

# for item in feats:
#     print(item)
print(len(feats))

print(len(custom_mapping))
```

```python
# Dataframe of coefficients
coefs = pd.DataFrame(model.coef_, columns=feats, index=custom_mapping.keys())

# In column names, strip 'num__' and 'cat__'
coefs.columns = coefs.columns.str.replace('num__', '').str.replace('cat__', '')

# Output to Excel
coefs.to_excel('../../../../Output/Modelling/Logistic Regression/' + model_name + '/' + model_name + '_coefs.xlsx')

coefs
```
