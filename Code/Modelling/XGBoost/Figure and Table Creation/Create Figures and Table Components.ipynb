# Create Figures and Table Components

```python
# Packages
import sys
import os
# Add the parent directory to sys.path
sys.path.append(os.path.abspath('..'))
# Import module
from XGBoost_Functions import *
```

## Set Classifier Name and Models to iterate over

```python
# Rating model names
include_exclude_previous = ['exclude_previous_', 'include_previous_']
model_names = ['rating_model_1', 'rating_model_2', 'rating_model_3']
rating_model_names = []
for include_exclude in include_exclude_previous:
    for model_name in model_names:
        rating_model_names.append(include_exclude + model_name)

# Change model names
prefixes = ['rating_change_', 'smote_rating_change_']
endings = ['model_1', 'model_2', 'model_3']
change_model_names = []
for prefix in prefixes:
    for ending in endings:
        change_model_names.append(prefix + ending)

# Full model names
full_model_names = rating_model_names + change_model_names
```

## Iterate Over Full Model Names and Create Figures and Tables

```python
for full_model_name in full_model_names:
    # Get the column names and mapping
    if not 'change' in full_model_name:
        _, _, target_column, custom_mapping = get_column_names_and_mapping(full_model_name)
    else:
        _, _, target_column, custom_mapping = get_column_names_and_mapping_change(full_model_name)
    print('creating figures and tables for ' + full_model_name)
    create_model_figure_and_table_components(full_model_name, target_column, custom_mapping)
    print('done creating figures and tables for ' + full_model_name)
```
