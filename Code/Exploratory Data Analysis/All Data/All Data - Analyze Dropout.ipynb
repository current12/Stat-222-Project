## Load and setup

```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data
df = pd.read_parquet(r'~\Box\STAT 222 Capstone\Intermediate Data\All_Data\all_data_fixed_quarter_dates.parquet') 
df
```

## Compute max date and look for cases where it isn't 2016-10-01, supposed last quarter

```python
# For each ticker, get the max value of fixed_quarter_date, and print out items where it is not in the last quarter of 2016
# Convert fixed_quarter_date to datetime
df['fixed_quarter_date_dt'] = pd.to_datetime(df['fixed_quarter_date'])
# Add column max_date to df
df['max_date'] = df.groupby('ticker')['fixed_quarter_date_dt'].transform('max')
# Display rows where max_date is not in the last quarter of 2016
df[df['max_date'] < '2016-10-01']
```

## Firms dropping out and last date appearing

```python
# Unique firms where this is the case
print(len(df[df['max_date'] < '2016-10-01'][['ticker', 'max_date']].drop_duplicates()))
print('share of firms dropping out:', len(df[df['max_date'] < '2016-10-01'][['ticker', 'max_date']].drop_duplicates()) / df['ticker'].nunique())
dropout_firms = df[df['max_date'] < '2016-10-01'][['ticker', 'max_date']].drop_duplicates().sort_values(by='max_date')
dropout_firms
```

Most are close which can be OK due to date inaccuracy

## Full quarter history for these firms

Output to Excel for investigation

```python
dropout_firms_full_hist = df[df['ticker'].isin(dropout_firms['ticker'])][['ticker', 'fixed_quarter_date', 'earnings_call_date', 'calendarYear', 'period']].sort_values(by=['ticker', 'fixed_quarter_date'])
dropout_firms_full_hist.to_excel(r'dropout_firms_full_hist.xlsx', index=False)
dropout_firms_full_hist
```

## Check bankruptcies

validate with bankruptcies dataset

```python
# Load all Excel files in the folder "~\Box\STAT 222 Capstone\Raw Data\Bankruptcies" and concatenate
import os
import glob
path = r'C:\Users\ijyli\Box\STAT 222 Capstone\Raw Data\Bankruptcies'
all_files = glob.glob(os.path.join(path, "*.xlsx"))
print(all_files)
df_from_each_file = [pd.read_excel(f) for f in all_files]
df_bankruptcies = pd.concat(df_from_each_file, ignore_index=True)
df_bankruptcies['in_bankruptcy'] = 1
df_bankruptcies
```

```python
# companies with credit rating data
# "~\Box\STAT 222 Capstone\Intermediate Data\Credit_Rating\combined_credit_rating_data.csv"
combined_credit = pd.read_csv(r'~\Box\STAT 222 Capstone\Intermediate Data\Credit_Rating\combined_credit_rating_data.csv')
combined_credit
```

```python
# combined credit symbols
combined_credit_symbol_df = combined_credit[['Symbol']].drop_duplicates()
combined_credit_symbol_df['in_credit'] = 1
combined_credit_symbol_df
```

```python
# Outer join df_bankruptcies and combined_credit on symbol
df_bankruptcies_credit = df_bankruptcies.merge(combined_credit_symbol_df, how='outer', left_on='Symbol', right_on='Symbol')
# Fill in_credit and in_bankruptcy with 0
df_bankruptcies_credit['in_credit'] = df_bankruptcies_credit['in_credit'].fillna(0)
df_bankruptcies_credit['in_bankruptcy'] = df_bankruptcies_credit['in_bankruptcy'].fillna(0)
# Counts of in_credit and in_bankruptcy
print(df_bankruptcies_credit[['in_credit', 'in_bankruptcy']].value_counts())
```
