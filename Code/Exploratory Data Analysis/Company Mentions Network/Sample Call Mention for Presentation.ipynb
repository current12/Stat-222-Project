# Sample Call Mention for Presentation

Find a company mention to use an example in the presentation

```python
# Packages
import pandas as pd
import os
```

## Load Mentions

```python
mentions = pd.read_excel('../../../Data/Company_Mentions/Company_Mentions_With_Ticker.xlsx')
mentions
```

```python
# Try Netflix's mentions
netflix = mentions[mentions['ticker'] == 'NFLX']
netflix
```

```python
# Filter for a specific mention
co_to_find = 'GOOGLE'
specific_mention = netflix[netflix['company_mentioned'] == co_to_find]
specific_mention
```

## Load Call

```python
file_list = [f for f in os.listdir(r'../../../Data/All_Data/All_Data_with_NLP_Features') if f.endswith('.parquet')]
# read in all parquet files
df = pd.concat([pd.read_parquet(r'../../../Data/All_Data/All_Data_with_NLP_Features/' + f, columns = ['ticker', 'fixed_quarter_date', 'transcript']) for f in file_list])
# convert fixed_quarter_date to string
df['fixed_quarter_date'] = df['fixed_quarter_date'].astype(str)
df
```

```python
call_text = df.merge(specific_mention, on = ['ticker', 'fixed_quarter_date'])['transcript'][0]
print(call_text)
```
